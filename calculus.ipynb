# Calculus

Let's start with a little trick
We draw a line starting at a with a gradient of a and a line starting at b with gradient b.

```python
x = array(range(12))
fig,ax =plt.subplots()
axis([0,10,0,10])
ax.set_yticks([2,3,6])
ax.set_yticklabels(["1*2","3*1","3*2"])
ax.set_xticks([1,2,3,5])
ax.set_xticklabels(["1","2","3=2+1","5=3+2"])
def line(n):
    plot(x,np.add(np.multiply(x,n),-n**2))
    
line(1)
line(2)
line(3)
ax.grid(True)
```

Something funny happens, the lines intersect at the points $(a*b,a+b)$.

It's easy to see.
The lines are described by

$a(a-x) = ax-a^2.$

So an intersection of two lines is

$ax-a²=bx-b²$

Now we plug in $(a+b)$ for x which gives us 

$a(a+b)-a² = b(a+b)-b² = a*b$

We can also solve for x which gives us 

$ x=(a²-b²)/(a-b) $
also called the differential quotient.

Now as the diffence between a and b becomes smaller and eventually zero , they seem to become the the tangents for a parabola.

```python
fig,ax =plt.subplots()
x = array(range(100))
ax.grid(True)
ax.axis([0,40,0,100])
fig.set_size_inches(20,10)
for i in np.arange(1,13):
    #plot(x,(np.multiply(x,i)))
    plot(x,(np.add(np.multiply(x,i),-i**2)))

#axis([0,100,0,100])
plot(x,(x/2)**2)
```

So let's figure it out by using algebra.

\begin{equation}
a=b
\end{equation}

\begin{equation}
x=\frac{a²-b²}{a-b}
\end{equation}

Well we should not divide by zero, but we can apply the binomial rule

\begin{equation}
a²-b²=(a+b) \cdot (a-b)
\end{equation}
So:

\begin{equation}
\frac{a²-b²}{a-b}=a+b
\end{equation}
So now with $a=b$ we get

\begin{equation}
a+a=2a.
\end{equation}
Going back to $ax-a²$ with $x=2a$ it becomes

\begin{equation}
2a²-a² = a²
\end{equation}

Which is how we conclude $f'(x²) = 2x$.

So let's derive some calculus rules with it:

#### Addition rule:

\begin{equation}
\frac{f(a)+g(a)-f(b)+g(b)}{a-b}=\frac{f(a)-f(b)}{a-b}+\frac{g(a)-g(b)}{a-b}
\end{equation}

#### Chain rule:

\begin{equation}
\frac{f(g(a))-f(g(b))} {a-b} = \frac{f(g(a))-f(g(b))}{g(a)-g(b)} \cdot \frac{(g(a)-g(b))}{a-b}
\end{equation}

#### Product rule:

\begin{equation}
\frac {f(a)g(a)-f(b)g(b)} {a-b} = 
\frac{g(b)f(a)}{a-b}-\frac{g(b)f(b)}{a-b}-\frac{f(a)f(b)}{a-b}-\frac{f(a)g(b)}{a-b} 
= g(b) \cdot \frac{f(a)-f(b)}{a-b} + f(a) \cdot \frac{(g(a)-g(b))}{a-b}
\end{equation}

#### Powers:

\begin{equation}
\frac{a^{2}-b^{2}}{a-b}=a+b = 2a
\end{equation}

\begin{equation}
\frac{a^{3}-b^{3}}{a-b}=a^{2}+ab+b^{2} = 3a^{2}
\end{equation}

\begin{equation}
\frac{a^{4}-b^{4}}{a-b}=a^{3}+a^{2}b+ab^{2}+b^{3} = 4a^{3}
\end{equation}

\begin{equation}
\frac {a^{n}-b^{n}}{a-b} = a^{n-1}+a^{n-2}b^{1}+...+b^{n-2}a^{1}+b^{n-1} = na^{n-1}
\end{equation}
